### Q1 

```python
def isPalindrome(s:str) -> bool:
    strs = []
    for char in s:
        if char.isalnum():
            strs.append(char.lower())

    while len(strs) > 1 :
        if strs.pop(0) != strs.pop():
            return False
    return True
```

```python
def isPalindrome_Deque(s:str) -> bool:
    # 자료형 데크로 선언
    strs: Deque = collections.deque()
    for char in s:
        if char.isalnum():
            strs.append(char.lower())

    while len(strs) > 1 :
        if strs.popleft() != strs.pop():
            return False
    return True
```

```python
def isPalindrome_regex(s:str) -> bool:
    s = s.lower()
    # 정규식으로 불필요한 문자 필터링
    s = re.sub('[^a-z0-9]','',s)
    
    return s == s[::-1]
```

### Q3

```python
paragraph = "Bob hit a ball, the hit BALL flew far after it was hit. flew. ball, flew, flew, flew"
banned = ["hit"]
```

```python
words = [word for word in re.sub(r'[^\w]', ' ', paragraph).lower().split() if word not in banned]
```

```python
counts = collections.defaultdict(int)
for word in words:
    counts[word] += 1
```

```python
max(counts, key=counts.get)
```
